```python
fig, ax = plt.subplots()
upper_x = 10.
upper_y = 1.
xs = np.linspace(1.,upper_x,100)
def boundary(x):
    return 1./(1.+x)
ys = map(boundary,xs)
upper_ys = np.ones(len(xs))*100
not_unimodal = matplotlib.patches.Rectangle((0,0),width=upper_x,height=upper_y,color='r',label='not unimodal',zorder=1)
ax.add_patch(not_unimodal)
ax.fill_between(xs,ys,upper_ys,color='b',zorder=10)
rect = matplotlib.patches.Rectangle( (0,0), width=0, height=0,color='b',label='unimodal')
ax.add_patch(rect)
ax.legend(prop={'size':constants.legend_fontsize}).set_zorder(10)
ax.set_xlim((0,upper_x))
ax.set_ylim((0,upper_y))
ax.plot([1.,upper_x],[upper_y,upper_y],linestyle='--')
ax.plot([upper_x,upper_x],[boundary(upper_x),upper_y],linestyle='--')
ax.set_xlabel('$B$',fontsize=constants.label_fontsize)
ax.set_ylabel('$\mu$',rotation=0,fontsize=constants.label_fontsize)
ax.set_title('Regions of unimodality for $\operatorname{Beta}(\mu,B)$ distribution',fontsize=constants.title_fontsize)
ax.tick_params(axis='both', which='major', labelsize=constants.tick_fontsize)
ax.tick_params(axis='both', which='minor', labelsize=constants.tick_fontsize)
fig.tight_layout()
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
not_unimodal = matplotlib.patches.Rectangle((1,0),width=upper_x-1,height=upper_y,color='r',label='not unimodal',zorder=1)
ax.add_patch(not_unimodal)
unimodal = matplotlib.patches.Rectangle((0,0),width=1,height=upper_y,color='b',label='unimodal',zorder=1)
ax.add_patch(unimodal)
ax.set_xlim((0,upper_x))
ax.set_ylim((0,upper_y))
ax.legend(prop={'size':constants.legend_fontsize}).set_zorder(10)
ax.set_xlabel('$\phi$',fontsize=constants.label_fontsize)
ax.set_ylabel('$m$',rotation=0,fontsize=constants.label_fontsize)
ax.set_title('Regions of unimodality for $\operatorname{Beta}(m,\phi)$ distribution', fontsize=constants.title_fontsize)
ax.tick_params(axis='both', which='major', labelsize=constants.tick_fontsize)
ax.tick_params(axis='both', which='minor', labelsize=constants.tick_fontsize)
fig.tight_layout()
basic.display_fig_inline(fig)
```

```python
import matplotlib.pyplot as plt
import matplotlib
from IPython.display import display_pretty, display_html
import python_utils.python_utils.basic as basic
import recovery_curve.recovery_curve.utils as recovery_utils
import recovery_curve.recovery_curve.fxns as recovery_fxns
import numpy as np
import functools
import pdb
```

```python
# define the distribution used for inference
s_a, s_b, s_c = 1., 1., 1.
l_a, l_b, l_c = 10., 10., 10.
K = 1
l_a_dist = recovery_utils.frozen_dist(functools.partial(recovery_utils.sample_truncated_exponential,l=l_a))
l_b_dist = recovery_utils.frozen_dist(functools.partial(recovery_utils.sample_truncated_exponential,l=l_b))
l_c_dist = recovery_utils.frozen_dist(functools.partial(recovery_utils.sample_truncated_exponential,l=l_c))
B_phis_dist = recovery_fxns.B_phis_dist(s_a, s_b, s_c, l_a_dist, l_b_dist, l_c_dist, K)
pop_a, pop_b, pop_c = 0.4, 0.7, 5.
a_dist = recovery_fxns.a_dist(pop_a)
b_dist = recovery_fxns.b_dist(pop_b)
c_dist = recovery_fxns.c_dist(pop_c)
abc_ns_dist = recovery_fxns.abc_ns_dist(a_dist, b_dist, c_dist)
l_m = 10.
noise_dist_dist = recovery_fxns.noise_dist_dist(l_m)
everything_dist = recovery_fxns.everything_dist(B_phis_dist, abc_ns_dist, noise_dist_dist)
```

```python
# define the distribution used for simulating data, and simulate it
N = 5000
stdev = 1.
x_ns = recovery_fxns.simulate_xs(N, K, stdev)
ts_ns = [np.linspace(1,50,10) for n in xrange(N)]
s_ns = np.ones(N)
b_a = 1.
b_b = 2.
b_c = 3.
B_a = b_a * np.ones(K)
B_b = b_b * np.ones(K)
B_c = b_c * np.ones(K)
phi_a, phi_b, phi_c = 0.01, 0.01, 0.01
constant_B_phis_dist = recovery_utils.constant_dist(recovery_fxns.B_phis(B_a, B_b, B_c, phi_a, phi_b, phi_c))
constant_abc_ns_dist = abc_ns_dist # this means abc_ns_dist does not contain within it a source of randomness, ie f(x;g); g is random
theta = 0.1
p = 0.3
phi_m = 0.01
constant_noise_dist_dist = recovery_utils.constant_dist(recovery_fxns.noise_dist(theta,p,phi_m))
constant_everything_dist = recovery_fxns.everything_dist(constant_B_phis_dist, constant_abc_ns_dist, constant_noise_dist_dist)
everything_sample = constant_everything_dist.sample(s_ns, x_ns, ts_ns)
ys_ns_sample = everything_sample.ys_ns
```

```python
# get posterior samples
n_steps = 2500
random_seed = 42
num_chains = 4
traces = recovery_fxns.get_everything_dist_traces_helper(n_steps, random_seed, everything_dist, s_ns, x_ns, ts_ns, ys_ns_sample, num_chains)
```

```python
# plot histograms of B_phis

fig, ax = plt.subplots()
traces.print_diagnostic('B_a', ax)
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
traces.print_diagnostic('B_b', ax)
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
traces.print_diagnostic('B_c', ax)
fig.suptitle('B_c')
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
traces.print_diagnostic('phi_a', ax)
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
traces.print_diagnostic('phi_b', ax)
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
traces.print_diagnostic('phi_c', ax)
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
traces.print_diagnostic('phi_m', ax)
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
traces.print_diagnostic('theta', ax)
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
traces.print_diagnostic('p', ax)
basic.display_fig_inline(fig)
```

```python
# plot box plot of distributions (B_a,B_b,B_c,p,theta), (phi_a,phi_b,phi_c,phi_m)
fig = plt.figure()
ax = fig.add_subplot(1,2,1)
data = [\
        traces.param_to_trace('B_a'),\
        traces.param_to_trace('B_b'),\
        traces.param_to_trace('B_c'),\
        traces.param_to_trace('theta'),\
        ]
true_vals = [\
             b_a,\
             b_b,\
             b_c,\
             theta,\
             ]
labels = [\
          '$B_a$',\
          '$B_b$',\
          '$B_c$',\
          '$\Theta$'
          ]
ax.boxplot(data)
ax.set_xticklabels(labels)
for (i,true_val) in enumerate(true_vals):
    ax.plot([i+1-0.25,i+1+0.25],[true_val,true_val],lw=3,color='k')

ax = fig.add_subplot(1,2,2)

data = [\
        traces.param_to_trace('phi_a'),\
        traces.param_to_trace('phi_b'),\
        traces.param_to_trace('phi_c'),\
        traces.param_to_trace('phi_m'),\
        ]
true_vals = [\
             phi_a,\
             phi_b,\
             phi_c,\
             phi_m,\
             ]
labels = [\
          '$\phi_a$',\
          '$\phi_b$',\
          '$\phi_c$',\
          '$\phi_m$'
          ]
ax.boxplot(data)
ax.set_xticklabels(labels)
for (i,true_val) in enumerate(true_vals):
    ax.plot([i+1-0.25,i+1+0.25],[true_val,true_val],lw=3,color='k')

basic.display_fig_inline(fig)
```

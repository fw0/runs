```python
# define the recovery fitter used for inference
s_a, s_b, s_c = 1., 1., 1.
l_a, l_b, l_c = 10., 10., 10.
#l_a_dist = recovery_utils.frozen_dist(functools.partial(recovery_utils.sample_truncated_exponential,l_a))
#l_b_dist = recovery_utils.frozen_dist(functools.partial(recovery_utils.sample_truncated_exponential,l_b))
#l_c_dist = recovery_utils.frozen_dist(functools.partial(recovery_utils.sample_truncated_exponential,l_c))
l_a_dist = recovery_utils.constant_dist(0.3)
l_b_dist = recovery_utils.constant_dist(0.3)
l_c_dist = recovery_utils.constant_dist(0.8)
B_phis_dist_fitter = recovery_fxns.B_phis_dist_fitter(s_a, s_b, s_c, \
                                                      l_a_dist, l_b_dist, l_c_dist)
l_m = 10.
noise_dist_dist = recovery_fxns.noise_dist_dist(l_m)
everything_dist_fitter = recovery_fxns.everything_dist_fitter(B_phis_dist_fitter, noise_dist_dist)
n_steps = 1000
random_seed = 42
traces_helper = recovery_fxns.get_everything_with_test_dist_traces_helper
point_f = np.mean
recovery_fitter = recovery_fxns.recovery_fitter(n_steps, random_seed, everything_dist_fitter, traces_helper, point_f)
```

```python
# get the prior distribution
everything_dist = everything_dist_fitter(recovery_X, ys_ns)
```

```python
# figure out the indices of distinct patients in dataset, allocate the figures
s = set()
idxs = []
for (i,x_i) in enumerate(recovery_X.x_ns):
    if x_i.sum() not in s:
        idxs.append(i)
        s.add(x_i.sum())
patient_figs = [plt.figure() for i in xrange(len(idxs))]
print idxs
```

```python
# for each distinct patient, draw samples from their prior and plot them
N_samples = 100
s_ns, x_ns, ts_ns = recovery_X
x_ns_display = recovery_X.x_ns[idxs]
ts_ns_display = np.linspace(0,50,51)
s_display = 1.
for (x_i,fig) in itertools.izip(x_ns_display,patient_figs):
    ax = fig.add_subplot(1,2,1)
    for i in xrange(N_samples):
        B_phis_sample, noise_dist_sample, abc_ns_sample, ys_ns_sample = everything_dist.sample([s_display],[x_i],[ts_ns_display])
        a_sample,b_sample,c_sample = abc_ns_sample[0]
        ax.plot(ts_ns_display,np.vectorize(functools.partial(recovery_utils.f,s_display,a_sample,b_sample,c_sample))(ts_ns_display),color='b',lw=0.5,alpha=0.25)
        ax.set_title('prior',fontsize=constants.title_fontsize)
        ax.set_xlabel('time (months)',fontsize=constants.label_fontsize)
        ax.set_ylabel('scaled function value',fontsize=constants.label_fontsize)
```

```python
# get posterior traces
n_steps = 100
random_seed = 42
s_ns, x_ns, ts_ns = recovery_X
traces = recovery_fxns.get_everything_with_test_phis_fixed_dist_traces_helper(n_steps, random_seed, everything_dist, s_ns, x_ns, ts_ns, ys_ns, s_ns, x_ns)
```

```python
# for the patients of interest, plot their curves
a_ns_trace = traces.param_to_trace('as_test')
b_ns_trace = traces.param_to_trace('bs_test')
c_ns_trace = traces.param_to_trace('cs_test')
for (i,fig) in itertools.izip(idxs,patient_figs):
    ax = fig.add_subplot(1,2,2)
    for (a_sample,b_sample,c_sample) in itertools.izip(a_ns_trace[:,i],b_ns_trace[:,i],c_ns_trace[:,i]):
        ax.plot(ts_ns_display,np.vectorize(functools.partial(recovery_utils.f,s_display,a_sample,b_sample,c_sample))(ts_ns_display),color='b',lw=0.5,alpha=0.25)
    ax.set_title('posterior',fontsize=constants.title_fontsize)
    ax.set_xlabel('time (months)',fontsize=constants.label_fontsize)
    ax.set_ylabel('scaled function value',fontsize=constants.label_fontsize)
    ax.tick_params(axis='both', which='major', labelsize=constants.tick_fontsize)
    ax.tick_params(axis='both', which='minor', labelsize=constants.tick_fontsize)
    fig.tight_layout()
    basic.display_fig_inline(fig)
```

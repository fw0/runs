```python
import numpy as np
np.random.seed(42)
import python_utils.python_utils.caching as caching
import recovery_run.recovery_run.constants as constants
caching.init(constants.cache_folder, constants.which_hash)
import matplotlib
matplotlib.use('agg')
```

```python
from IPython.display import display_pretty, display_html
import python_utils.python_utils.basic as basic
import matplotlib.pyplot as plt
import recovery_curve.recovery_curve.utils as recovery_utils
import recovery_curve.recovery_curve.fxns as recovery_fxns
import numpy as np
import functools, itertools
```

```python
# define constants
alpha = 0.05
curve_max_lim = 20
xs = np.linspace(0,curve_max_lim,100)
n_samples = 1000
color = 'k'
actual_S = 0.9
```

```python
# first plot
fake_S = 0.5
A_dist = recovery_utils.beta_mode_phi_dist(0.3, 0.02)
B_dist = recovery_utils.beta_mode_phi_dist(0.8, 0.05)
C_dist = recovery_utils.gamma_mode_phi_dist(5.0, 0.01)
A_samples = [A_dist.sample() for i in xrange(n_samples)]
B_samples = [B_dist.sample() for i in xrange(n_samples)]
C_samples = [C_dist.sample() for i in xrange(n_samples)]
fig, ax = plt.subplots()
for (A,B,C) in itertools.izip(A_samples, B_samples, C_samples):
    f = functools.partial(recovery_utils.f, actual_S, A, B, C)
    ax.plot(xs, map(f, xs), alpha=alpha, color=color,lw=1)
ax.plot([constants.curve_neg_lim,0],[fake_S,fake_S], color=color,lw=constants.lw)
ax.axvline(0, ymin=0, ymax=fake_S, color=color, lw=1)
ax.set_xlim((constants.curve_neg_lim))
ax.set_xlim(constants.curve_neg_lim,curve_max_lim)
ax.set_ylim((0,1))
ax.set_xlabel('months',fontsize=constants.label_fontsize)
ax.set_ylabel('fxn value',fontsize=constants.label_fontsize)
ax.tick_params(axis='both', which='major', labelsize=constants.tick_fontsize)
ax.tick_params(axis='both', which='minor', labelsize=constants.tick_fontsize)
fig.tight_layout()
basic.display_fig_inline(fig)
fake_S, alpha
```

```python
from IPython.display import display_pretty, display_html
import python_utils.python_utils.basic as basic
import matplotlib.pyplot as plt
import recovery_curve.recovery_curve.utils as recovery_utils
import recovery_curve.recovery_curve.fxns as recovery_fxns
import numpy as np
```

```python
# define some shared variables
n_samples = 1000
n_bins = 20
beta_xs = np.linspace(0,1,101)
gamma_xs = np.linspace(0,25,250)
phis = [.001,.005,.01,.1,.2]
```

```python
# simulate from a gamma distribution, and plot histogram and pdf
mode, phi = 5., 0.01
n_samples = 1000
gamma_dist = recovery_utils.gamma_mode_phi_dist(mode, phi)

fig, ax = plt.subplots()
ax.hist([gamma_dist.sample() for i in xrange(n_samples)],bins=n_bins)
ax.set_title('samples from gamma(%.2f,%.2f) distribution' % (mode,phi), fontsize=constants.title_fontsize)
basic.display_fig_inline(fig)

fig, ax = plt.subplots()
ax.plot(gamma_xs, np.exp(map(gamma_dist.loglik, gamma_xs)))
ax.set_title('logpdf of gamma(%.2f,%.2f) distribution' % (mode,phi), fontsize=constants.title_fontsize)
basic.display_fig_inline(fig)
```

```python
# for fixed mode, plot many gamma distributions
mode = 5.
fig, ax = plt.subplots()
for phi in phis:
    gamma_dist = recovery_utils.gamma_mode_phi_dist(mode, phi)
    ax.plot(gamma_xs, np.exp(map(gamma_dist.loglik, gamma_xs)), label='$\phi\'=%.3f$' % phi)
ax.set_title(r'pdfs of $\operatorname{gamma}_{m,\phi}(m\',\phi\')$ distributions with $m\'=%d$' % mode, fontsize=constants.title_fontsize)
ax.set_xlabel('x', fontsize=constants.label_fontsize)
ax.set_ylabel('p(x)', fontsize=constants.label_fontsize)
ax.legend(prop={'size':constants.legend_fontsize})
ax.tick_params(axis='both', which='major', labelsize=constants.tick_fontsize)
ax.tick_params(axis='both', which='minor', labelsize=constants.tick_fontsize)
fig.tight_layout()
basic.display_fig_inline(fig)
```

```python
# helpers for plotting

def plot_pdf(ax, f, label=None, maxlim=1):
    xs = np.linspace(0,maxlim,100)
    ys = map(f, xs)
    ax.plot(xs, ys, label=label)
    
curve_max_x = 20
S=0.9
left=-2
def plot_curve(ax, f):
    xs = np.linspace(0,curve_max_x,200)
    ys = map(f, xs)
    ax.plot((0,0),(f(0),S),lw=1,color='k',alpha=0.5)
#    ax.plot((0,0),(f(0),0.5),lw=1,color='k',alpha=0.5)
#    ax.plot((left,0),(0.5,0.5),lw=3,color='k',alpha=1)
    ax.plot((left,0),(S,S),lw=5,color='k',alpha=1)
    ax.plot(xs, ys, color='k', alpha=0.02, lw=2)
```

```python
# plot unimodal beta pdfs, vary mode
phi = 0.01
modes = [.1,.2,.3,.4,.5]
fig, ax = plt.subplots()
for mode in modes:
    import pdb
    logpdf = recovery_utils.beta_mode_phi_dist(mode,phi).loglik
    pdf = lambda x: np.exp(logpdf(x))
    plot_pdf(ax, pdf)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
basic.display_fig_inline(fig)
```

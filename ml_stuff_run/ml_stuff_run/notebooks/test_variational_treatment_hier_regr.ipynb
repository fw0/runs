```python
# random seed
np.random.seed(41)
```

```python
# define params for simulating data
L = 3
N = 15000
d = 2
x_var = 15.0
mu = np.tile(5.0, d)
#prec_val = 100000
prec_val = 1
lambda_val = 1
delta = np.cumsum(np.ones(L))
treatment_prob = 0.4
```

```python
# simulate the data
x_ns, y_ns, z_ns, T_ns, B_ls = hvr.simulate_data(L, N, d, x_var, mu, prec_val, lambda_val, delta, treatment_prob)
```

```python
# define prior by specifying hyperparameters
v_0 = 1.0001
assert v_0 > d-1
T_0 = 1.0 * np.eye(d) 
c_0 = .0001 
m_0 = np.zeros(d)
alpha_0 = 1.
beta_0 = 1.
r_0 = -1000
s_0 = 1000
```

```python
# define params for doing inference
num_iters = 5
```

```python
# initialize params
v, T, c, m, mu_B_ls, prec_B_ls, alpha_lambda_ls, beta_lambda_ls, mu_delta, prec_delta = hvr.init_variational_params\
(x_ns, y_ns, z_ns, T_ns, v_0, T_0, c_0, m_0, alpha_0, beta_0, r_0, s_0)
```

```python
# do inference
v, T, c, m, mu_B_ls, prec_B_ls, alpha_lambda_ls, beta_lambda_ls, mu_delta, prec_delta = hvr.infer_variational_params(num_iters, x_ns, y_ns, z_ns, T_ns, v_0, T_0, c_0, m_0, alpha_0, beta_0, r_0, s_0,\
                         v, T, c, m, mu_B_ls, prec_B_ls, alpha_lambda_ls, beta_lambda_ls, mu_delta, prec_delta)
```
